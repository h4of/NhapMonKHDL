```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

from sklearn.preprocessing import OneHotEncoder
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
```

```python
df_car = pd.read_csv('car_clean (1).csv')
df_car.head()
```

```python
df_car['Giá'].value_counts()
```

```python
train_df, test_df = train_test_split(df_car, test_size=0.2, random_state=42)
```

```python
num_cols = ['Năm SX','Km đã đi']
categorical_cols = ['Nhiên liệu', 'Kiểu dáng', 'Tình trạng', 'Hộp số', 'Xuất xứ','Tên','Tỉnh thành']
target_col = 'Giá'
```

```python
scaler = StandardScaler()
train_df.head()
X_train_num = scaler.fit_transform(train_df[num_cols])
X_test_num = scaler.transform(test_df[num_cols])

encoder = OneHotEncoder(sparse_output=False, drop='first', handle_unknown='ignore')
X_train_cat = encoder.fit_transform(train_df[categorical_cols])
X_test_cat = encoder.transform(test_df[categorical_cols])

X_train_full = np.concatenate([X_train_num, X_train_cat], axis=1)
X_test_full = np.concatenate([X_test_num, X_test_cat], axis=1)

# encoded_df_train = pd.DataFrame(X_train_full, columns=np.concatenate((num_cols, encoder.get_feature_names_out(categorical_cols))))
# encoded_df_test = pd.DataFrame(X_test_full, columns=np.concatenate((num_cols, encoder.get_feature_names_out(categorical_cols))))

# encoded_df_train
```

```python
model = LinearRegression()
model.fit(X_train_full, train_df[target_col])

y_pred = model.predict(X_test_full)
mse = mean_squared_error(test_df[target_col], y_pred)
r2 = r2_score(test_df[target_col], y_pred)
mae = mean_absolute_error(test_df[target_col], y_pred)

print("Mean Squared Error:", mse)
print("Mean Absolute Error", mae)
print("R-squared:", r2)
```
