```python
import numpy as np
import pandas as pd
from time import sleep
from selenium import webdriver
import random
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
```

```python
def safe_find_text(driver,by, selector, replace_text="\n"):
    try:
        return driver.find_element(by, selector).text.replace(replace_text, "")
    except:
        return "N/A"
```

```python
driver = webdriver.Chrome(service=Service("chromedriver.exe"))

TT=['Năm SX','Nhiên liệu','Kiểu dáng','Tình trạng','Km đã đi','Hộp số','Xuất xứ','Tỉnh thành']

index = 1
Xe=[]
while True:
    driver.get(f"https://oto.com.vn/mua-ban-xe/p{index}")
    sleep(random.randint(1,3))
    elements = driver.find_elements(By.CSS_SELECTOR, ".item-car .info .info-left .title [href]")
    links = [el.get_attribute('href') for el in elements]
    for link in links:
      driver.get(link)
      sleep(random.randint(1,3))
      TT_xe = {}
      TT_xe['Ngày đăng bài'] = safe_find_text(driver,By.XPATH,"/html/body/div[2]/div[2]/div[2]/div[1]/div[5]/div/div[2]/span[1]")
      TT_xe['Tên'] =  safe_find_text(driver,By.CSS_SELECTOR,".group-title-detail")
      TT_xe['Giá'] = safe_find_text(driver,By.CSS_SELECTOR,".box-price .price-big")
      
      Tinh_trang_xe=[]
      for i in range(1,3):
        for j in range(1,5):
          Tinh_trang_xe.append(safe_find_text(driver,By.XPATH, f"/html/body/div[2]/div[2]/div[2]/div[1]/div[3]/div[2]/ul[{i}]/li[{j}]"))
      Tinh_trang_xe = [tuple(tt.split(":",1)) for tt in Tinh_trang_xe if ":" in tt]
      for tt in TT:
        try:
          TT_xe[tt] = dict(Tinh_trang_xe)[tt]
        except:
          TT_xe[tt] = "N/A"
      Xe.append(TT_xe)
    
    index+=1
    if index == 140 :
        break


```

```python
df = pd.DataFrame(Xe)
```
